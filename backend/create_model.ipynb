```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
train = pd.read_csv('data\ViCTSD_train.csv')
X_train = train['Comment']
y_train = train['Toxicity']
```

```python
train
```

```python
test = pd.read_csv('data\ViCTSD_test.csv')
X_test = test['Comment']
y_test = test['Toxicity']
```

```python
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
from sklearn.inspection import permutation_importance
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
# Huấn luyện mô hình SVM
svm_model = SVC(kernel='rbf', C=1.2)
svm_model.fit(X_train_tfidf, y_train)
svm_predictions = svm_model.predict(X_test_tfidf)

# Convert the sparse matrix to a dense NumPy array
X_test_dense = X_test_tfidf.toarray()

# perm_importance = permutation_importance(svm_model, X_test_dense, y_test) # Use the dense array here
# features = np.array(feature_names)
# sorted_idx = perm_importance.importances_mean.argsort()
# plt.barh(features[sorted_idx], perm_importance.importances_mean[sorted_idx])
# plt.xlabel("Permutation Importance")

# Đánh giá mô hình SVM
print("SVM Classification Report:")
print(classification_report(y_test, svm_predictions))
print("SVM Accuracy:", accuracy_score(y_test, svm_predictions))
```

```python
import pickle
import os

# Tạo thư mục 'model' nếu chưa tồn tại
os.makedirs('model', exist_ok=True)

# Giả sử bạn đã có svm_model và vectorizer

# Lưu mô hình SVM
with open('model/svm_model.pkl', 'wb') as file:
    pickle.dump(svm_model, file)

# Lưu vectorizer
with open('model/vectorizer.pkl', 'wb') as file:
    pickle.dump(vectorizer, file)
```

```python
model = pickle.loads(open('model/svm_model.pkl', 'rb').read())
vectorizer = pickle.loads(open('model/vectorizer.pkl', 'rb').read())
```
