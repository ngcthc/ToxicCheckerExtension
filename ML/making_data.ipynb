```python
import pandas as pd
```

```python
train = pd.read_csv('data/down_train.csv')
valid = pd.read_csv('data/down_valid.csv')
test = pd.read_csv('data/down_test.csv')
VOZ = pd.read_csv('data/down_VOZ.csv')
```

```python
from sklearn.model_selection import train_test_split

train_VOZ, valid_VOZ = train_test_split(VOZ, test_size=0.2, random_state=42)
```

```python
train = pd.concat([train, train_VOZ])
valid = pd.concat([valid, valid_VOZ])
train
```

```python
train.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
valid.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
train.to_csv('data/train.csv', index=False)
valid.to_csv('data/valid.csv', index=False)
```
